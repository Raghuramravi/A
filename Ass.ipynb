```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV from GitHub
url = "https://raw.githubusercontent.com/Raghuramravi/b/main/clean_data.csv"
df = pd.read_csv(url)

# Show sample of player info
display(df[['name', 'country', 'height', 'weight', 'position_number']].head())

# Group by country to see number of players
players_by_country = df['country'].value_counts().reset_index()
players_by_country.columns = ['Country', 'Number_of_Players']

# Display summary table
display(players_by_country.head())

# Plot top 10 countries
plt.figure(figsize=(10,6))
sns.barplot(data=players_by_country.head(10), x='Number_of_Players', y='Country', palette='cubehelix')
plt.title("üåç Top 10 Countries by Number of Players")
plt.xlabel("Number of Players")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
